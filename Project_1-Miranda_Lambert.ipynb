```python
def histogram_plotter(title, data):
    sns.set(rc={"figure.figsize": (6, 4)})
    sns.set_style('white')
#     dist = sns.distplot(data, kde=False, bins=30)
    dist = sns.distplot(data, hist_kws={'alpha':0.2}, kde_kws={'linewidth':5})
    
    dist.axvline(np.mean(data), lw=5, c='darkred', ls='dashed')
    average_roll = '{0:.2f}'.format(np.mean(data))
    title = 'Distribution of ' + title + '; average = '  + average_roll + '\n'
    dist.set_title(title, fontsize=16)
    dist = sns.distplot(data, hist_kws={'alpha':0.2}, kde_kws={'linewidth':5})
```

Analysis Report


I. Introduction 
   
   The data presented here was collected over a four month period in the summer and fall of 2016 in Medellin, Colombia with the goal of assessing gender differences in competitiveness. 192 parents from six schools in the two lowest income strata participated in one of fourteen sessions. Each participant played a total of six games and then took a demographic survey. The first four games require parents to answer as many addition questions (up to 20) possible within 3 minutes. In the first round, players are paid a set rate per correct answer. The second game is played the same except players are forced to play a random and unknown opponent with the winner being paid double the round one reward if they win and zero if they lose. Round three and four are our main rounds of interest for this project and are referred to as "choice cash" and "choice voucher"  respectively. In these rounds, players must choose whether to play with the lower per correct answer payment method or choose choose to compete with the tournament round payment style and get twice as much or nothing. The difference in the two rounds is the means of payment. In round 3, participants are paid in cash and round four the payment is a voucher which their child can use to get snacks at their school tienda. The final two games illicit measures of risk and willingness to pay. The original hypothesis is that men will prove more competitive than women for cash, but women will become almost if not equally ompetitive for a voucher benefitting their child directly. 
   
II. Summary Statistics and Data Cleaning    
   
   The current analysis goes through basic summary statistics, correlations, and finally data visualizations in order to assess the difference in willingness to compete (WTC)between genders. Because this is a repliacation of Cassar, Whordofa, and Zhang, 2015, after a quick data frame visualization, a mask is created to drop all non-parent participants from the analysis for consistency. Next, columns with all missing values are removed, empty values are replaced with the mean in important columns, and variable names are changed for clarity. Grouping the data by gender allows us to see any differences between males and females. Here we find a significant difference in the tournament round score between genders with males scoring significantly higher. This could play an important part in ones decision to compete. 
   
III. Correlations   
   
   Correlations allow for the understanding of how variables relate to one another. Not surprisingly, the heat map shows choice to compete for cash (choice0) is highly correlated to choice to compete for voucher (choice1). Here we can see once again evidence of a strong negative correlation between score in the tournament round and gender. Additionally, score is positively related to both someone's choice to compete in each round and with their income. Finally, the heat map shows some positive relationship between forced displacement and a person's choice to compete, which we will further test in the next section. 
   The next section of the analysis uses group by functions to understand the core questions of this research. Here we find males to be slightly more willing to compete than females for cash and the child-benefitting voucher, but a more in depth analysis is needed to determine whether this difference is significant (previous analysis in fact shows that the gender difference is not significant). In an effort to understand women's high rates of competitiveness in this study, the analysis continues by looking at the effect of forced displacement on competitiveness. Women who have been forcefully displaced prove to be much more likely to compete than those who have not, while men compete at similar rates regardless of displacement status. 
   
IV. Graphical Representations
   
   This section begins with a box plot to test for outliers in one of the few continuous variables, age. Here we see a few outliers on the upper end, but due to data size constraints choose not to drop them. In addition, a histogram shows age to be well distributerd despite the outliers. Similarly, histograms of score and confidence show their distribution within the sample. Somewhat concerningly, we see a build up of low scoring individuals with between 0 and 5 correct answers, but this is likely still an accurate representation of the sample population as a whole and therefore not too concerning. 
   The nice part about working with experimental data is its inherently random nature. Because the entire collection process is randomized, simple graphical visualizations can tell the full story. The graphs entitled "WTC for Cash by Gender" and "WTC for Voucher by Gender" paint a similar representation as to what is going on in this study. We see that men and women are almost equally likely to compete regardless of the prize for which they are competing. To me the most interesting part of this analysis is in the displacement analysis. The graphs entitled "WTC for Cash/Voucher by Displacement" show those who have been displaced to be slightly more likely to compete. However, separating this by gender shows that women are single-handedly driving this difference. In particular, we see that mothers who have been forcefully displaced compete at much higher levels for the child-benefitting voucher.
