```python
##All code from 4 made in colaboration with Daniel Salicath and Thoedore Wisinski.  

##fixing the date column and creating quarters

iowa['Date']=pd.to_datetime(iowa['Date'])
iowa ['year'], iowa['month']=iowa['Date'].dt.year, iowa['Date'].dt.month
iowa['quarter']=iowa ['Date'].dt.quarter
iowa.head()
```

```python
##This creates aggregate sales for each quarter by individual store locations

quarter_sales_group = iowa.groupby(['store','year', 'quarter'])['sales', 'state_retail', 'state_cost'].agg(np.sum).reset_index()
quarter_sales_group.rename(columns = {'sales': 'quarterly_sales'}, inplace=True)

quarter_sales_group.head(5)
```

```python
iowa_merge=iowa.merge(quarter_sales_group, on=['store', 'year', 'quarter'], how='left')
iowa_merge.head()
```

```python
##creating dataframes for each quarter and then separating quarters for of 2015


quarter1_sales=iowa[iowa['quarter']==1].groupby(['store', 'year'])['sales'].agg(np.sum).reset_index()
quarter1_sales=quarter1_sales.rename(columns={'sales': 'quarter1_sales'})
quarter1_sales.head()
```

```python
quarter1_sales_15=quarter1_sales[quarter1_sales['year']==2015].groupby(['store'])['quarter1_sales'].agg(np.sum).reset_index()
quarter1_sales_15=quarter1_sales_15.rename(columns={'quarter1_sales':'quarter1_sales15'})
quarter1_sales15=iowa.merge(quarter1_sales_15, on= ['store'], how='left')
quarter1_sales15.head()
```

```python
##quarter 2
quarter2_sales=iowa[iowa['quarter']==2].groupby(['store', 'year'])['sales'].agg(np.sum).reset_index()
quarter2_sales=quarter2_sales.rename(columns={'sales':'quarter2_sales'})
quarter2_sales.head()
```

```python
##quarter 2 of 2015
quarter2_sales15=quarter2_sales[quarter2_sales['year']==2015].groupby(['store'])['quarter2_sales'].agg(np.sum).reset_index()
quarter2_sales15=quarter2_sales15.rename(columns={'quarter2_sales':'quarter2_sales15'})
##quarters 1 and 2
quarter12_sales15=quarter1_sales15.merge(quarter2_sales15, on=['store'], how='left')

quarter12_sales15.head()
```

```python
##quarter 3
quarter3_sales=iowa[iowa['quarter']==3].groupby(['store', 'year'])['sales'].agg(np.sum).reset_index()
quarter3_sales=quarter3_sales.rename(columns={'sales':'quarter3_sales'})
quarter3_sales15=quarter3_sales[quarter3_sales['year']==2015].groupby(['store'])['quarter3_sales'].agg(np.sum).reset_index()
quarter3_sales15=quarter3_sales15.rename(columns={'quarter3_sales':'quarter3_sales15'})
##combine quarters 1,2,and 3
quarter123_sales15=quarter12_sales15.merge(quarter3_sales15, on=['store'], how='left')
quarter123_sales15.head()
```

```python
##quarter 4 
quarter4_sales=iowa[iowa['quarter']==4].groupby(['store', 'year'])['sales'].agg(np.sum).reset_index()
quarter4_sales=quarter4_sales.rename(columns={'sales':'quarter4_sales'})
quarter4_sales15=quarter4_sales[quarter4_sales['year']==2015].groupby(['store'])['quarter4_sales'].agg(np.sum).reset_index()
quarter4_sales15=quarter4_sales15.rename(columns={'quarter4_sales':'quarter4_sales15'})
##All 4 2015 quarters
quarterly_sales15=quarter123_sales15.merge(quarter4_sales15, on=['store'], how='left')
quarterly_sales15.head()
```

```python
##First quarter of 2016
quarter1_sales16=quarter1_sales[quarter1_sales['year']==2016].groupby(['store'])['quarter1_sales'].agg(np.sum).reset_index()
quarter1_sales16=quarter1_sales16.rename(columns={'quarter1_sales':'quarter1_sales16'})
quarter1_sales16.head()
```

```python
quarter_sales=quarterly_sales15.merge(quarter1_sales16, on=['store'], how='left')
quarter_sales.head()
```

```python
quarter_sales['sales_2015']=quarter_sales['quarter1_sales15']+quarter_sales['quarter2_sales15']+quarter_sales['quarter3_sales15']+quarter_sales['quarter4_sales15']
quarter_sales.sales_2015.head()

```

```python
##replace missing values with 0 to see who was not open in the quarters

def missing(d):
    if d.isnull().sum()>0:
        print 'replaced=', d.isnull().sum()
        d.fillna(0, inplace=True)
    elif d.isnull().sum() == 0:
        print 'no missing'
        
missing(quarter_sales.quarter1_sales15)
missing(quarter_sales.quarter2_sales15)
missing(quarter_sales.quarter3_sales15)
missing(quarter_sales.quarter4_sales15)
missing(quarter_sales.quarter1_sales16)
```

```python
##Clean the data of all stores that had a full quarter with no sales. This eludes to the fact that they either opened or closed within the year.

mask_q1=quarter_sales['quarter1_sales15']!=0
quarter1_sales_mask=quarter_sales[mask_q1].reset_index()

mask_q12=quarter1_sales_mask['quarter2_sales15']!=0
quarter12_sales_mask=quarter1_sales_mask[mask_q12].reset_index()

mask_q123=quarter12_sales_mask['quarter3_sales15']!=0
quarter123_sales_mask=quarter12_sales_mask[mask_q123]

mask_q1234=quarter123_sales_mask['quarter4_sales15']!=0
quarter1234_sales_mask=quarter123_sales_mask[mask_q1234]

mask_q12345=quarter1234_sales_mask['quarter1_sales16']!=0
quarter_sales=quarter1234_sales_mask[mask_q12345]
quarter_sales.describe()
```
